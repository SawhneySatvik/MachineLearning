## Simple Logistic Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification # For generating binary classification data
```

```python
X, y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=422)
```

```python
X.shape, y.shape
```

```python
X = pd.DataFrame(X)
y = pd.Series(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=422)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python
log_reg.fit(X_train, y_train)
```

```python
y_pred = log_reg.predict(X_test)
```

```python
y_pred_proba = log_reg.predict_proba(X_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, classification_report
```

```python
confusion_matrix_res = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(confusion_matrix_res, display_labels=log_reg.classes_).plot()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
acc_score = accuracy_score(y_test, y_pred)
pre_score = precision_score(y_test, y_pred)
rec_score = recall_score(y_test, y_pred)
f_score = f1_score(y_test, y_pred)
acc_score, pre_score, rec_score, f_score
```
